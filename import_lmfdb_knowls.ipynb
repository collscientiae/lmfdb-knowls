```python
for knowl in knowls.find(fields=["_id", "authors", "content", "timestamp", "title"]):
    ids = knowl["_id"].split(".")
    if len(ids) > 1:
        path = sep.join(ids[:-1])
        if not exists(path):
            makedirs(path)
    fn = sep.join(ids) + ".md"
    with open(fn, "wb", "utf8") as out:
        print "writing", knowl["_id"], "to", fn
        if "title" in knowl:
            out.write("title: %s\n" % knowl["title"])
        else:
            out.write("title: %s\n" % knowl["_id"])
        if "authors" in knowl:
            out.write("authors:\n")
            for a in knowl["authors"]:
                out.write('    %s\n' % a)
        #if "timestamp" in knowl:
        #    out.write("date: %s\n" % knowl["timestamp"])
        out.write("\n")
        out.write(knowl["content"])
        
```

# Post Processing

```python
from os import walk, unlink
from os.path import sep, splitext
import jinja2 as j2
```

```python
j2env = j2.Environment(loader=j2.FileSystemLoader("."), undefined=j2.StrictUndefined)
def url_for(*args, **kwargs):
    return "<b>%s</b>" % args[0]
j2env.globals["url_for"] = url_for
j2env.globals["LINK_EXT"] = lambda a, b: "<a href='%s'>%s</a>" % (a, b)
j2env.globals["TEXT_DATA"] = lambda a, b: "TEXT_DATA: %s/%s" % (a,b)
j2env.globals["group_alias_table"] = lambda : "<code>group_alias_table()</code>"
```

```python
macros = """\
{% macro KNOWL(kid, title=none) -%}
<a knowl="lmfdb/{{ kid }}">{{ title or kid }}</a>
{%- endmacro %}

"""
```

```python
def no_blanks_on_top(string):
    ret = []
    ontop = True
    for line in string.splitlines():
        if ontop and len(line) == 0:
            continue
        ontop = False
        ret.append(line)
    return "\n".join(ret)
```

```python
for path, _, files in walk("."):
    for fn in files:        
        if fn in [".git", "README.md"]:
            continue
        fn_split = splitext(fn)
        if len(fn_split) != 2 or fn_split[1] != ".md":
            continue
        fn2 = fn
        if fn == "index.md":
            print "fn is 'index', renaming to 'index_'"
            fn2 = "index_.md"
        fn = sep.join([path, fn])
        fn2 = sep.join([path, fn2])
        print fn, "=>", fn2
        with open(fn, "r", "utf8") as input:
            content = input.read()
            content = macros + content
        try:
            rendered = j2env.from_string(content).render()
            rendered = no_blanks_on_top(rendered)
            with open(fn2, "wb", "utf8") as out:
                out.write(rendered)
                out.write("\n")
            if fn != fn2:
                unlink(fn)
        except Exception as e:
            print e
            print content
            break
```
